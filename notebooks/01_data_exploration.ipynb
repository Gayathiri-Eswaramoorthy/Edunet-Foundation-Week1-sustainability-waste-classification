```python
# --- Importing Libraries ---
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Import ImageDataGenerator with graceful fallback if TensorFlow is not available
try:
    from tensorflow.keras.preprocessing.image import ImageDataGenerator
except Exception as e:
    try:
        from keras.preprocessing.image import ImageDataGenerator
    except Exception:
        ImageDataGenerator = None
        print("ImageDataGenerator import failed. Install TensorFlow or Keras to proceed. Error:", e)
```

```python
# --- Dataset Path ---
train_dir = "../data/train"
```

```python
# --- View Categories ---
categories = os.listdir(train_dir)
print("Waste Categories:", categories)
```

```python
# --- Display a Sample Image from Each Category ---
for category in categories:
    path = os.path.join(train_dir, category)
    sample_img = cv2.imread(os.path.join(path, os.listdir(path)[0]))
    plt.imshow(cv2.cvtColor(sample_img, cv2.COLOR_BGR2RGB))
    plt.title(category)
    plt.axis('off')
    plt.show()
```
